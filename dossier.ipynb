### Code multiplication matrice * vecteur python
puis nous avons commencé notre implémentatiion du probléme avec la création de fonction qui permet d'effectuer les action importante du probléme. ici la multiplication d'une matrice et d'un vecteur sans utilisé de bibliothéque suplémentaire.

ce morceau de code permet de multiplier une matrice et un vecteur tant que leur nombre de le ligne est le même. puis la fonction retourne un vecteur de même taille que celui d'entré qui contient le résulta de la multiplication.

```python
def multiplyMatrixVector(myMatrix, myVector):
    resultVector = [0]*len(myVector)
    for i in range(len(myMatrix)):
        result = 0
        for j in range(len(myMatrix[i])):
            result += myMatrix[i][j] * myVector[j]
        resultVector[i] = result
    return resultVector
```

### Code somme des vecteur 


Cette fonction , prend en paramettre un vecteur et retourne la somme de tout les valeurs qu'elle contient.
Cette fonction nous permet de vérifier la somme des lignes, colones et des diagonales d'une matrice, pour observer comment ce comporte la population quand les ligne, colones ou diagonal de la matrice change. 

```python
def sumVector(myVector):
    result = 0
    for i in range(len(myVector)):
        result += myVector[i]
    return result
```
