### Code multiplication matrice * vecteur python
puis nous avons commencé notre implémentatiion du probléme avec la création de fonction qui permet d'effectuer les action importante du probléme. ici la multiplication d'une matrice et d'un vecteur sans utilisé de bibliothéque suplémentaire.

ce morceau de code permet de multiplier une matrice et un vecteur tant que leur nombre de le ligne est le même. puis la fonction retourne un vecteur de même taille que celui d'entré qui contient le résulta de la multiplication.

```python
def multiplyMatrixVector(myMatrix, myVector):
    resultVector = [0]*len(myVector)
    for i in range(len(myMatrix)):
        result = 0
        for j in range(len(myMatrix[i])):
            result += myMatrix[i][j] * myVector[j]
        resultVector[i] = result
    return resultVector
```

### Code somme des vecteur 


Cette fonction , prend en paramettre un vecteur et retourne la somme de tout les valeurs qu'elle contient.
Cette fonction nous permet de vérifier la somme des lignes, colones et des diagonales d'une matrice, pour observer comment ce comporte la population quand les ligne, colones ou diagonal de la matrice change. 

```python
def sumVector(myVector):
    result = 0
    for i in range(len(myVector)):
        result += myVector[i]
    return result
```

### Pour générer les matrices de test nous utilisons différentes fonctions :

Tout d’abord nous avons genMatricLineEqual(n,x) qui nous génère un matrice n² dons la somme de chaque ligne est égale à x, cette fonction fait aussi que la somme de la colonne 1 est plus grande que la somme de la colonne 2 et ainsi de suite.

```python
import random as rd
def genMatricLineEqual(n,x):
    mat = []
    for i in range(n):
        ligne = []
        borne_sup=float(x)
        diff = 0.0
        for j in range(n):
            if j == n-1 :
                ligne.append(round(float(x)-diff,2))
            else :
                tmp = round(rd.uniform(0.0, borne_sup),2)
                ligne.append(tmp)
                borne_sup -= tmp 
                diff += tmp
        mat.append(ligne)

    return mat

print(genMatricLineEqual(5,1))
```

Nous utilisons également genMatricColEqual(n,x) qui nous génère un matrice n² dons la somme de chaque colonne est égale à x, cette fonction fait aussi que la somme de la ligne 1 est plus grande que la somme de la ligne 2 et ainsi de suite.

```python
def genMatricColEqual(n,x):
    mat =[]
    diff = [0]*n
    borne_sup=[float(x)]*n

    for i in range(n):
        ligne = []
        for j in range(n):
            if i == n-1 :
                ligne.append(round(float(x)-diff[j],2))
            else :
                tmp = round(rd.uniform(0.0, borne_sup[j]),2)
                ligne.append(tmp)
                borne_sup[j] -= tmp 
                diff[j] += tmp
        mat.append(ligne)

    return mat

print(genMatricColEqual(4,3))
```

Et pour finir nous utilisons genMatricDiagEqual(n,x,reste) qui nous génère un matrice n² dons la somme de la diagonale principale est égale à x, pour les autres valeurs sont soit égale a 0 si la variable reste est égale à 0 et si la variable reste n’est pas égale à 0 les autres valeurs seront aléatoire.

```python
def genMatricDiagEqual(n,x,reste):
    mat =[]
    diff = 0
    borne_sup=float(x)

    for i in range(n):
        ligne = []
        for j in range(n):
            if i == n-1 and j ==n-1 :
                ligne.append(round(float(x)-diff,2))
            elif i==j :
                tmp = round(rd.uniform(0.0, borne_sup),2)
                ligne.append(tmp)
                borne_sup -= tmp 
                diff += tmp
            elif reste == 0 :
                ligne.append(0.0)
            else :
                ligne.append(round(rd.uniform(0.0, float(x)),2))
        mat.append(ligne)

    return mat

print(genMatricDiagEqual(5, 1, 0))
```
